```python
import requests
from bs4 import BeautifulSoup

page = requests.get("https://stackoverflow.com?tab=interesting")

if page.status_code == 200:
    soup = BeautifulSoup(page.content, 'html.parser')
    html = list(soup.children)[2]
    questions = soup.find_all('div', class_='question-summary narrow')
    
    for question in questions:
        print('Q : ' + question.select('div.summary h3 a.question-hyperlink')[0].get_text())
        print('Votes : ' + question.select('div.cp div.votes div.mini-counts span')[0].get_text())
        print('Answers : ' + question.select('div.cp div.status div.mini-counts span')[0].get_text())
        print('Views : ' + question.select('div.cp div.views div.mini-counts span')[0].get_text())
else:
    print('Couldn\'t read the page')
```
